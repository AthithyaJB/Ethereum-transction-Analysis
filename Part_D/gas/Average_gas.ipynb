```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Aver_gas.txt' ,header = None)
```

```python
df.head
```

```python
df.columns = ['date', 'gas']
```

```python
df.head(5)
```

```python
df['date'] = df['date'].str.strip('()')
df['date'] = df['date'].str.replace('"','')
df['gas'] = df['gas'].str.strip('()')
df['gas'] = df['gas'].astype(float)
df.head()
```

```python
df
```

```python
plt.figure(figsize=(25,15))
plt.title('Transaction Gas Price vs Time')
plt.xlabel('Date')
plt.ylabel('Average Gas')
plt.yticks(rotation=90)
plt.xticks(rotation=90)
plt.plot(df['date'], df['gas'])
plt.show()
```
